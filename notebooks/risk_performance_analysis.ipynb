```python
# Definimos nuestros activos de interés
activos = ['AAPL', 'MSFT', 'GOOGL', 'SPY', 'GLD'] # Tecnología, S&P500 y Oro

# Definimos el rango de tiempo (últimos 3 años para tener buen histórico)
fin = datetime.datetime.now()
inicio = fin - datetime.timedelta(days=3*365)

# Descargamos los datos
# Usamos ['Adj Close'] porque es el estándar para análisis de rendimiento
df_precios = yf.download(activos, start=inicio, end=fin, auto_adjust=False)['Adj Close']

# Revisamos las primeras filas
print(df_precios.head())
```

```python
# Descargamos los datos con auto_adjust=False para que aparezca la columna 'Adj Close'
df_precios = yf.download(activos, start=inicio, end=fin, auto_adjust=False)['Adj Close']

# Si el código anterior te da error de nuevo, usa este que es más robusto:
# data = yf.download(activos, start=inicio, end=fin)
# df_precios = data['Adj Close']

print(df_precios.head())
```

```python
# Calculamos el cambio porcentual diario
# .pct_change() compara el precio de hoy contra el de ayer
retornos = df_precios.pct_change().dropna()

# Visualizamos los retornos
print("Primeros retornos diarios:")
print(retornos.head())
```

```python
import numpy as np

# Calculamos la desviación estándar (volatilidad diaria)
volatilidad_diaria = retornos.std()

# Anualizamos la volatilidad
volatilidad_anual = volatilidad_diaria * np.sqrt(252)

# Mostramos el riesgo de cada activo (ordenado de mayor a menor)
print("\nVolatilidad Anualizada (Riesgo):")
print(volatilidad_anual.sort_values(ascending=False))
```

```python
retorno_total = (1 + retornos).prod() - 1
```

```python
import plotly.express as px

# Calculamos el retorno acumulado (Base 1)
# Esto nos dice: "¿Qué pasó con 1 dólar invertido al inicio?"
retorno_acumulado = (1 + retornos).cumprod()

# Creamos el gráfico interactivo
fig = px.line(retorno_acumulado,
              title='Crecimiento de $1 Invertido (Rendimiento Acumulado)',
              labels={'value': 'Valor de la Inversión', 'Date': 'Fecha'},
              template='plotly_dark') # Estilo elegante para finanzas

fig.show()
```

```python
import plotly.figure_factory as ff

# Calculamos la matriz de correlación
corr_matrix = retornos.corr().round(2)

# Creamos el mapa de calor interactivo
fig_corr = ff.create_annotated_heatmap(
    z=corr_matrix.values,
    x=list(corr_matrix.columns),
    y=list(corr_matrix.index),
    colorscale='Viridis',
    showscale=True
)

fig_corr.update_layout(title_text='Matriz de Correlación de Activos')
fig_corr.show()
```

```python
# Definimos el nivel de confianza (95%)
confianza = 0.05

# Calculamos el VaR histórico para cada activo
# El VaR es el percentil 5 de los retornos
var_historico = retornos.quantile(confianza)

print("Value at Risk (95%) Diario por activo:")
print(var_historico)

# Ejemplo de interpretación para tu portafolio:
# Si el VaR de Google es -0.025, significa que hay un 5% de probabilidad
# de perder más del 2.5% en un solo día.
```
